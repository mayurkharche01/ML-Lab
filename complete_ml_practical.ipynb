```

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder, MinMaxScaler, StandardScaler, PolynomialFeatures
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LogisticRegression, LinearRegression, Ridge, Lasso
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, mean_squared_error, r2_score

df = sns.load_dataset('titanic')
df = df[['survived','pclass','sex','age','fare','embarked']]

df.head()
```

## 1-2 Missing Values Handling

```

print(df.isnull().sum())

# Numerical
df['age'].fillna(df['age'].mean(), inplace=True)

# Categorical
df['embarked'].fillna(df['embarked'].mode()[0], inplace=True)

print(df.isnull().sum())
```

## 3 Label Encoding

```

le = LabelEncoder()
df['sex'] = le.fit_transform(df['sex'])
df.head()
```

## 4 One Hot Encoding

```

df = pd.get_dummies(df, columns=['embarked'])
df.head()
```

## 5 Min-Max Scaling

```

scaler = MinMaxScaler()
df[['age','fare']] = scaler.fit_transform(df[['age','fare']])
df[['age','fare']].describe()
```

## 6 Standard Scaling

```

sc = StandardScaler()
scaled = sc.fit_transform(df[['age','fare']])
print("Mean:", scaled.mean(), "Std:", scaled.std())
```

## 7 Feature Creation

```

df['fare_per_age'] = df['fare'] / (df['age']+1)
df.head()
```

## Prepare Data

```

X = df.drop('survived', axis=1)
y = df['survived']

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

## 15-20 Logistic Regression

```

model = LogisticRegression(max_iter=200)
model.fit(X_train,y_train)
y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

## 21 Scaling Comparison

```

sc = StandardScaler()
X_train_s = sc.fit_transform(X_train)
X_test_s = sc.transform(X_test)

model2 = LogisticRegression(max_iter=200)
model2.fit(X_train_s,y_train)
print("Scaled Accuracy:", accuracy_score(y_test,model2.predict(X_test_s)))
```

## 22-24 kNN

```

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train,y_train)
print("k=3 Accuracy:", accuracy_score(y_test,knn.predict(X_test)))

acc = []
for k in range(1,11):
    m = KNeighborsClassifier(n_neighbors=k)
    m.fit(X_train,y_train)
    acc.append(accuracy_score(y_test,m.predict(X_test)))

print(acc)
```

## 25 kNN with Scaling

```

knn.fit(X_train_s,y_train)
print("Scaled kNN:", accuracy_score(y_test,knn.predict(X_test_s)))
```

## 26-29 Linear Regression

```

lr = LinearRegression()
lr.fit(X_train,y_train)
pred = lr.predict(X_test)

print("MSE:", mean_squared_error(y_test,pred))
print("R2:", r2_score(y_test,pred))
```

## 30 Polynomial Regression

```

poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X_train)
lr.fit(X_poly,y_train)
```

## 31 Plot

```

plt.scatter(y_test,pred)
plt.show()
```

## 32 Linear Regression MSE

```

print(mean_squared_error(y_test,pred))
```

## 33-35 Ridge & Lasso

```

ridge = Ridge(alpha=1)
lasso = Lasso(alpha=0.1)

ridge.fit(X_train,y_train)
lasso.fit(X_train,y_train)

print("Ridge MSE:", mean_squared_error(y_test,ridge.predict(X_test)))
print("Lasso MSE:", mean_squared_error(y_test,lasso.predict(X_test)))
```

## 36 Cross Validation

```

scores = cross_val_score(lr,X,y,cv=5,scoring='neg_mean_squared_error')
print(scores.mean())
```

## 37-38 Best Model

```

print("Compare manually based on MSE above")
```

## 39-40 Alpha Tuning

```

alphas = [0.01,0.1,1,10]
errors = []
for a in alphas:
    r = Ridge(alpha=a)
    r.fit(X_train,y_train)
    errors.append(mean_squared_error(y_test,r.predict(X_test)))

plt.plot(alphas,errors)
plt.show()
```

## 41 Coefficients

```

print("Ridge:", ridge.coef_)
print("Lasso:", lasso.coef_)
```
